```python
# Install required libraries
!pip install diffusers transformers accelerate torch -q

# Import libraries
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Load Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

# Move model to GPU
pipe = pipe.to("cuda")

# Prompts
prompts = [
    "A futuristic smart city at sunset, ultra realistic",
    "A cyberpunk city at night with futuristic technology and neon lights",
    "An astronaut riding a horse on Mars",
    "A futuristic AI classroom with holograms",
    "A robot working in a modern office"
]

# Generate images
for i, prompt in enumerate(prompts):

    print(f"\nGenerating Image {i+1}")
    print("Prompt:", prompt)

    image = pipe(prompt).images[0]

    # Save image
    filename = f"generated_image_{i+1}.png"
    image.save(filename)

    # Show image
    plt.figure(figsize=(8,8))
    plt.imshow(image)
    plt.axis("off")
    plt.show()

print("Task completed successfully.")
```
